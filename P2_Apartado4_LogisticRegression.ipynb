# Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score
from matplotlib.colors import ListedColormap
```

# Vinos rojos

Se obtiene la información del archivo .csv y se almacena como un DataFrame

```python
df= pd.read_csv("winequality-red.csv",sep=';')
df.head()
```

Se dividen los datos del DataFrame en (X, y) 

```python
X = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
y = df['quality']
y.head()
```

Se dividen los datos en dos grupos, uno de entrenamiento y uno de prueba.

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.3, random_state=20)
```

Se estandarizan los valores de X para una correcta predicción.

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Se establece el modelo de regresión y se realiza la predicción de cada clase

```python
model = LogisticRegression(C=10**10, max_iter=100000)
model.fit(X_train, y_train.ravel())

y_pred = model.predict(X_test)
y_pred
```

Se comparan los valores de testeo con los valores predecidos

```python
np.c_[y_test, y_pred]
```

Se verifica qué tan preciso es nuestro modelo así como otros parámetros

```python
acc = accuracy_score(y_test.ravel(), y_pred)
print('Accuracy: %.3f' % acc)
```

```python
print(classification_report(y_test, y_pred))
```

Matriz de confusión

```python
confusion_matrix(y_test.ravel(), y_pred)
```

### Comparación entre la regresión lineal y la regresión logística

Regresión lineal

```python
model_linear = LinearRegression()
model_linear.fit(X_train, y_train.ravel())

y_pred_linear = model_linear.predict(X_test)
y_pred_linear_around = np.around(y_pred_linear)
y_pred_linear
```

```python
y_pred_linear_around
```

```python
np.c_[y_test, y_pred_linear_around]
```

```python
acc = accuracy_score(y_test.ravel(), y_pred_linear_around)
print('Accuracy: %.3f' % acc)
```

Se grafica la correlación entre el vector de salida de prueba y el vector de salida de la predicción. 
Regresión logística vs Regresión Lineal

```python
plt.figure(figsize=(16,12))

plt.subplot(2,2,1)
plt.scatter(y_pred, y_test, c=y_test, cmap='Spectral')
plt.title('Regresion Logística ')
plt.xlabel('Predicciones')
plt.ylabel('Valores de testeo')
plt.grid(True)

plt.subplot(2,2,4)
plt.scatter(y_pred_linear, y_test, c=y_test, cmap='Spectral')
plt.title('Regresion Lineal ')
plt.xlabel('Predicciones')
plt.ylabel('Valores de testeo')
plt.grid(True)

plt.subplot(2,2,2)
plt.scatter(y_pred_linear_around, y_test, c=y_test, cmap='Spectral')
plt.title('Regresion Lineal Redondeada')
plt.xlabel('Predicciones')
plt.ylabel('Valores de testeo')
plt.grid(True)
plt.show()
```

Se define una función para la generación de la gráfica de decision de las regiones

```python
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v', '*', 'd', 'p', '.')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan', 'yellow', 'orange', 'violet', 'pink')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2, = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                            np.arange(x2_min, x2_max, resolution))
    
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel(), xx1.ravel(), xx2.ravel(),
                                     xx1.ravel(), xx2.ravel(), xx1.ravel(), xx2.ravel(),
                                     xx1.ravel(), xx2.ravel(), xx1.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y.ravel() == cl, 0], y=X[y.ravel() == cl, 1],
                    alpha=0.8, c=colors[idx],
                    marker=markers[idx], label=cl,
                    edgecolor='black')
        
    # highlight test examples
    if test_idx:
        # plot all examples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0], X_test[:, 1],
                    c='', edgecolor='black', alpha=1.0,
                    linewidth=1, marker='o',
                    s=100, label='test set')
```

Gráfica de las regiones. Regresión logística vs Regresión lineal

```python
plt.figure(figsize=(10,10))

plt.subplot(2,1,1)
plot_decision_regions(X_test, y_test, classifier=model)
plt.legend()
plt.title('Regresión Logística')

plt.subplot(2,1,2)
plot_decision_regions(X_test, y_test, classifier=model_linear)
plt.legend()
plt.title('Regresión Líneal')

plt.show()
```

```python
confusion_matrix_linear = pd.crosstab(y_test.ravel(), y_pred_linear_around, rownames=['Valores de testeo'], 
                                      colnames=['Valores predecidos'])
confusion_matrix_logistic = pd.crosstab(y_test.ravel(), y_pred, rownames=['Valores de testeo'], 
                                        colnames=['Valores predecidos'])

plt.figure(figsize=(20,7))

plt.subplot(1,2,1)
sns.heatmap(confusion_matrix_logistic, annot=True)
plt.title('Regresión logística')

plt.subplot(1,2,2)
sns.heatmap(confusion_matrix_linear, annot=True)
plt.title('Regresión lineal redondeada')

plt.show()
```
